```python
import pandas as pd
```

### 1.Trazer a Base de dados

```python
tabela = pd.read_excel("Vendas.xlsx")
display(tabela)
```

### 2.Panorama geral da base de dados

```python
faturamento_total = tabela["Valor Final"].sum()
print(faturamento_total)
```

### 3.Análise Top -> Down

```python
# faturamento por loja
faturamento_por_loja = tabela[["ID Loja", "Valor Final"]].groupby("ID Loja").sum()
display(faturamento_por_loja)
```

##### Análise - Se a loja Iguatemi Campinas vende mais, o que pode ser que ela esteja fazendo de certo em relação as outras? talvez possa aplicar o mesmo para aumentar o faturamento das outras lojas

### 4.Entrar no detalhe para melhor entender se a hipótese faz sentido

```python
faturamento_por_produto = tabela[["ID Loja","Produto", "Valor Final"]].groupby(["ID Loja", "Produto"]).sum()
display(faturamento_por_produto)
```

### 5. Conclusão

##### Hipótese - Após análise foi notado que o item "Bermuda Liso" só está presente no shopping Iguatemi Campinas (que é onde mais possui faturamento), então uma possível hipótese é que ele é responsável pelo alto valor final e uma possível solução seria incluir o item nas outras lojas da empresa.
